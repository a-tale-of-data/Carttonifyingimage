```
from google.colab.patches import cv2_imshow
import cv2
```

```
img = cv2.imread('/content/sample_data/images/motor.png', 1)
cv2_imshow(img)
```

```
#converting to gray scale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
```

```
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
#We are using the bitwise_not function which is used to make brighter regions lighter and vice versa 
#so that we can find the edges to create a pencil sketch
```

```
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
```

```
def dodgeV2(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```
final_img = dodgeV2(img_gray, img_smoothing)
cv2_imshow(final_img)
```

```
filename = 'motor1.jpg'
cv2.imwrite(filename, final_img)
```
